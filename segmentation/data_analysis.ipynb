```python
import numpy as np
import matplotlib.pyplot as plt
home = "/home/localssk23/Downloads/privacy_instaseg/data"
```

```python
# Helper function to load data with memory mapping
def load_npy_file(filepath):
    return np.load(filepath, mmap_mode="r")
```

```python
# Official Splits
# Training: Fold 1; Validation: Fold 2; Testing: Fold 3
# Training: Fold 2; Validation: Fold 1; Testing: Fold 3
# Training: Fold 3; Validation: Fold 2; Testing: Fold 1
```

```python
fold = 3

img = load_npy_file(f"{home}/pannuke/Fold {fold}/images/fold{fold}/images.npy")
msk = load_npy_file(f"{home}/pannuke/Fold {fold}/masks/fold{fold}/masks.npy")

sample_img = img[0]
sample_msk = msk[0]
print()
print(f"Sample image shape: {sample_img.shape}")
print(f"Sample mask shape: {sample_msk.shape}")
print()

# plot all the 6 slices of the mask
fig, ax = plt.subplots(1, 6, figsize=(25, 5))
for i in range(6):
    ax[i].imshow(sample_msk[:, :, i])
plt.show()
```
